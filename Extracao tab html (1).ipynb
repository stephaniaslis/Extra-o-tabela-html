# Extração de tabelas de uma página html

## Dados: códigos de Unidades Federativas do IBGE


```python
import pandas as pd
```

```python
# Read html: o índice [0] significa que foi extraída d primeira coluna da página
UF = pd.read_html('https://www.ibge.gov.br/explica/codigos-dos-municipios.php')[0]
UF
```

```python
# Separando o código UF do restante
UF['Códigos'] = UF['Códigos'].str[:2]
UF['Códigos'].astype(int)
UF
```

```python
# Colocando o a sigla da UF corresponente ao seu nome
UF['UF_Sigla'] =["AC","AL","AP","AM","BA","CE","DF","ES","GO","MA","MT","MS","MG","PA","PB","PR","PE","PI","RN","RS","RJ","RO","RR","SC","SP","SE","TO"]
UF.sort_values(by = ['Códigos'], inplace = True)
UF
```
